```python
dxa_se = F.normalize(dxa_model.module.transformed_spatial_embeddings(sample['dxa_img'][None]),dim=1)
mri_se = F.normalize(mri_model.module.transformed_spatial_embeddings(sample['mri_img'][None]),dim=1)
```

```python
def get_4d_corr(se1, se2):
    b1, c1, h1, w1 = se1.size()
    b2, c2, h2, w2 = se2.size()
    corr = torch.bmm(se1.permute(0,2,3,1).view(b1,h1*w1,c1),se2.view(b2, c2, h2*w2))
    corr = corr.view(b1, h1, w1, h2, w2)
    return corr

def correlate_point(corr, pt=(30,30),tgt='dxa', softmax=True):
    assert tgt in ['dxa','mri']
    if tgt == 'dxa':
        corr_point_map = corr[:,pt[0],pt[1],:,:]
    else:
        corr_point_map = corr[:,:,:,pt[0],pt[1]]
    
    if softmax:
        b, h, w = corr_point_map.size()
        corr_point_map = F.softmax(corr_point_map.view(b,h*w),dim=1).view(b,h,w)
    
    
    return corr_point_map


def resample_point(src_img, tgt_img, point):
    _, hs, ws = src_img.size()
    _, ht, wt = tgt_img.size()
    resampled_point = [None, None]
    resampled_point[0] = np.round(point[0]*ht/hs).astype(np.int)
    resampled_point[1] = np.round(point[1]*wt/ws).astype(np.int)
    print(resampled_point)
    return resampled_point


img_point = (100, 138)
downsample_point = resample_point(dxa_vol, dxa_spatial_embeds[:,0],img_point)
corr = get_4d_corr(dxa_se, mri_se)

corr_point_map = correlate_point(corr, downsample_point)
plt.figure(figsize=(10,10))
plt.subplot(131)
plt.imshow(dxa_vol[0],cmap='gray')
plt.scatter(img_point[1], img_point[0],c='cyan',marker='x',s=100)
plt.subplot(132)
plt.imshow(mri_vol[0],cmap='gray')
plt.subplot(133)
plt.imshow(corr_point_map.detach()[0])
plt.show()





```
