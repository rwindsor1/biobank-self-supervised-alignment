```python
def align_ses(this_conv_ses, ses_shape, img_shape):
    ses_argmax = [int(this_conv_ses.argmax()//this_conv_ses.shape[-1]), int(this_conv_ses.argmax()%this_conv_ses.shape[-1])]
    img_dim = [ses_argmax[0]*img_shape[0]/ses_shape[0], ses_argmax[1]*img_shape[1]/ses_shape[1]]
    return img_dim
    
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
print(conv_ses.shape)
i=2
j=i

this_conv_ses = conv_ses[i,j]
plt.imshow(conv_ses[i,j])
img_dim = align_ses(this_conv_ses, dxa_ses.shape[2:], dxa_img.shape[2:])

plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(mri_img[j,0], cmap='gray')

plt.subplot(122)
plt.imshow(dxa_img[i,0], cmap='gray')
plt.gca().add_patch(Rectangle([img_dim[1],img_dim[0]], mri_img.shape[-1], mri_img.shape[-2],fc='none', ec='y'))

```

```python
def transform_ses(ses, pca):
    nc, ses_h, ses_w = ses.size()
    pca_ses = torch.tensor(pca.transform(ses.view(nc,-1).permute(1,0).numpy())).permute(1,0).view(3,ses_h, ses_w)
    return pca_ses

scan_idx = 5
    
# dxa only pca
dxa_only_pca = PCA(n_components=3)
dxa_only_pca.fit(all_dxa_ses.permute(1,0,2,3).reshape(128,-1).permute(1,0).numpy())
transformed_dxa_ses = transform_ses(dxa_ses[scan_idx],dxa_only_pca)

# mri only pca
mri_only_pca = PCA(n_components=3)
mri_only_pca.fit(all_mri_ses.permute(1,0,2,3).reshape(128,-1).permute(1,0).numpy())
transformed_mri_ses = transform_ses(mri_ses[scan_idx],mri_only_pca)
plt.figure(figsize=(8,15))
plt.subplot(221)
plt.imshow(dxa_img[scan_idx,0],cmap='gray')
plt.subplot(222)
plt.imshow(transformed_dxa_ses.permute(1,2,0))
plt.subplot(223)
plt.imshow(mri_img[scan_idx,0],cmap='gray')
plt.subplot(224)
plt.imshow(transformed_mri_ses.permute(1,2,0))
plt.show()

plt.imshow(transformed_mri_ses.permute(1,2,0).sum(dim=-1))
```

# View Correlation Maps, DXA -> MRI

```python
from gen_utils import red, grayscale
def get_corr(dxa_se, mri_se):
    bd, cd, hd, wd = dxa_se.size()
    bm, cm, hm, wm = mri_se.size()
    corr = torch.bmm(dxa_se.permute(0,2,3,1).view(bd,hd*wd,cd),mri_se.view(bm, cm, hm*wm))
    corr = corr.view(bd, hd, wd, hm, wm)
    return corr

def correlate_point(corr, pt=(30,30),tgt='dxa', softmax=True):
    assert tgt in ['dxa','mri']
    if tgt == 'dxa':
        corr_point_map = corr[:,pt[0],pt[1],:,:]
    else:
        corr_point_map = corr[:,:,:,pt[0],pt[1]]
    temperature = 0.1
    if softmax:
        b, h, w = corr_point_map.size()
        corr_point_map = F.softmax(corr_point_map.view(b,h*w)/temperature,dim=1).view(b,h,w)
    
    
    return corr_point_map


def resample_point(src_img, tgt_img, point):
    _, hs, ws = src_img.size()
    _, ht, wt = tgt_img.size()
    resampled_point = [None, None]
    resampled_point[0] = np.round(point[0]*ht/hs).astype(np.int)
    resampled_point[1] = np.round(point[1]*wt/ws).astype(np.int)
    return resampled_point

scan_idx=0
img_point = (300,100)
corr4d = get_corr(dxa_ses, mri_ses)

downsample_point = resample_point(dxa_img[:,0], dxa_ses[:,0],img_point)
corr_point_map = correlate_point(corr4d, downsample_point)

plt.figure(figsize=(10,20))
plt.subplot(131)
plt.title('Source DXA + Point')
plt.imshow(dxa_img[scan_idx,0], cmap='gray')
plt.scatter(img_point[1], img_point[0],marker='x',c='cyan',s=100)
plt.subplot(132)
plt.title('MRI Correlation Map')
plt.axis('off')
plt.imshow(corr_point_map[scan_idx])
plt.subplot(133)
plt.axis('off')
plt.title('Target MRI')

interpolated_corr_point_map = F.interpolate(corr_point_map[scan_idx][None,None], mri_img.shape[2:])[0,0]
plt.imshow(grayscale(mri_img[scan_idx,0])+ red((interpolated_corr_point_map-interpolated_corr_point_map.min())/
                                               (interpolated_corr_point_map.max()-interpolated_corr_point_map.min())))

plt.tight_layout()
```

# Try RANSAC



```python
def get_most_discriminative_points(points, corr_map, dxa_img, mri_img,mode='mri', num_points=80):
    scored_points = []
    for idx, point in enumerate(points):
        if mode=='dxa':src_img = dxa_img; tgt_img= corr_map[:,:,0,0][None];other_img = mri_img; 
        else: src_img = mri_img; tgt_img= corr_map[0,0,:,:][None]; other_img = dxa_img;
        ses_point = resample_point(src_img,tgt_img, point)
        corr_point_map = correlate_point(corr_map[None], ses_point, tgt=mode, softmax=True)
        tgt_ses_pt = [int(corr_point_map[0].argmax() // corr_point_map.shape[2]), int(corr_point_map[0].argmax() % corr_point_map.shape[2])]
        corres_point = resample_point(corr_point_map, other_img, tgt_ses_pt)
        scored_points.append([point, corres_point, corr_point_map.max(), ses_point])
        
    scored_points.sort(key=lambda x: -x[2])
    return [x[0] for x in scored_points[:num_points]],[x[1] for x in scored_points[:num_points]]
    
    
# def RANSAC(src_pts, tgt_pts):
    
dxa_scan_idx = 2; mri_scan_idx = 2
print(corr4d.shape)
X,Y =np.meshgrid(np.linspace(10,mri_img.shape[2]-10,40),np.linspace(10,mri_img.shape[3]-10,20))


corr4d = get_corr(dxa_ses[dxa_scan_idx][None], mri_ses[mri_scan_idx][None])
print(corr4d.shape)
points = np.stack((X.reshape(-1),Y.reshape(-1))).transpose()
points = points[np.random.permutation(np.arange(points.shape[0]))]
discrim_points,match_points = get_most_discriminative_points(points,corr4d[0],dxa_img[dxa_scan_idx], mri_img[mri_scan_idx])

print(dxa_img.shape)
mri_crp = np.concatenate([mri_img[mri_scan_idx,0], np.zeros((dxa_img.shape[-2]-mri_img.shape[-2],mri_img.shape[-1]))])
plt.figure(figsize=(15,15))

plt.imshow(grayscale(np.concatenate([mri_crp,dxa_img[dxa_scan_idx,0]],axis=1)))
for idx, point in enumerate(discrim_points):
    plt.scatter(point[1],point[0],color='y',marker='.')
    match_point = match_points[idx]
    plt.scatter(match_point[1]+mri_img.shape[-1],match_point[0],marker='.',c='y')
    plt.plot([match_point[1]+mri_img.shape[-1],point[1]],
             [match_point[0],point[0]])
plt.show()

```
